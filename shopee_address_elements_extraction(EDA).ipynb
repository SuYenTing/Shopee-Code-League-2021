# 處理POI資料不一致之狀況
### => 建立一個錯誤POI寫法與正確POI寫法的對照表

```
# 處理POI不一致之狀況
poiNotSameData = trainData[[x[0] not in x[1] for x in zip(trainData['poi'], trainData['raw_address'])]]
poiNotSameData = poiNotSameData.reset_index(drop=True)
poiNotSameData
```

```
# 自定義文字相似分數
def StrCompareScore(word1, word2):
  return sum([1 for x in zip(word1, word2) if x[0] == x[1]])  
```

```
# 處理poi與raw_address不一致之問題
# 紀錄資訊
errorPoi = list()
correctPoi = list()

# 迴圈各筆資料
for j in tqdm(range(len(poiNotSameData))):

  token1 = poiNotSameData['raw_address'][j]
  token1 = token1.replace(',', '')
  token2 = poiNotSameData['poi'][j]
  token2 = token2.replace(',', '')

  if token2:

    # 將文字切為token
    token1 = token1.split()
    token2 = token2.split()

    # 找出POI的位置
    scoreList = []
    for i in range(len(token1)-len(token2)+1):
      score = sum([StrCompareScore(x[0], x[1]) for x in zip(token1[i:(i+len(token2))], token2)])
      scoreList.append(score)
    matchSite = scoreList.index(max(scoreList))

    # 紀錄差異字
    token1target = token1[matchSite:(matchSite+len(token2))]
    token2target = token2
    if len([x for x in zip(token1target, token2target) if x[0] != x[1]]) > 0:
      errorPoi.append(token1target)
      correctPoi.append(token2target)
```

```
# 輸出資料
errorPoiTable = pd.DataFrame({'errorPoi': errorPoi, 'correctPoi': correctPoi})
errorPoiTable.to_csv(filePath + 'errorPoiTable.csv', index=False)
errorPoiTable
```
